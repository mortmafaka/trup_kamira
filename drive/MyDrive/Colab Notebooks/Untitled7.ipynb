```python
soup = BeautifulSoup(response.text, 'lxml')
scripts = soup.find_all('script')
for script in scripts:
        script_text = script.get_text()

        # Используем регулярное выражение для поиска текста после "return"
        matches = re.findall(r'return\s+(.+)', script_text)
        if matches:
            for match in matches:
                print("Found after 'return':", match)
```

```python
import re
list_of_json = []

text = script.get_text()

# Регулярное выражение для поиска JSON
pattern = r'\{.*?\}'  # для поиска объектов
# pattern = r'\[.*?\]'  # для поиска массивов

matches = re.findall(pattern, text)
for match in matches:
    print(match)
    list_of_json.append(match)
print(len(list_of_json))
```

```python
# Найдем все JSON-структуры, начинающиеся с {id: и заканчивающиеся на }
text = list_of_json
ids = []
for i in (text):
    if 'id:a' in i:
        ids.append(i)
```

```python
products = []
#headers = ['name','price','calories']
for i in ids:
  if 'name' in i and 'price' in i and 'calories' in i:
    products.append(i)

```

```python
extracted_data = []

for product in products:
    product_data = {}

    # Извлечение значения ключа 'name'
    name_match = re.search(r'name:"([^"]*)"', product)
    if name_match:
        product_data['name'] = name_match.group(1)

    # Извлечение значения ключа 'consist'
    consist_match = re.search(r'consist:"([^"]*)"', product)
    if consist_match:
        product_data['consist'] = consist_match.group(1)

    # Для 'info' пока просто проверяем наличие, так как это сложный объект
    # Если вам нужно извлекать конкретные подключи из 'info', дайте знать
    info_match = re.search(r'info:\{([^\}]*)\}', product)
    if info_match:
        # Это будет строкой, представляющей содержимое info.
        # Если вам нужно преобразовать это в другой словарь, дайте знать.
        product_data['info'] = info_match.group(1)

    if product_data:
        extracted_data.append(product_data)

# Теперь у вас есть список словарей extracted_data с извлеченной информацией
print(extracted_data)
```

```python
import csv

# ... ваш код, где вы создаете список extracted_data ...

# Создание файла CSV
with open('products.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['name', 'consist', 'info']  # Заголовки столбцов
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()  # Запись заголовков
    for data in extracted_data:
        writer.writerow(data)
```
