```python
from requests_html import HTMLSession
```

```python
url = 'https://www.ligastavok.ru/bets/my-line/soccer/rossiia-id-350/rossiiskaia-premer-liga-id-5271/spartak-moskva-zenit-spb-id-18281280'
```

```python
from requests_html import AsyncHTMLSession

async def fetch():
    session = AsyncHTMLSession()
    r = await session.get(url)
    await r.html.arender()
    print(r.html.raw_html)

# Запуск асинхронной функции
import asyncio
asyncio.run(fetch())
```
