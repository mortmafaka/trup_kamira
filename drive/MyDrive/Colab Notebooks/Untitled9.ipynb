```python
import requests

def verify_recaptcha(recaptcha_response, secret_key):
    # URL для проверки reCAPTCHA
    verify_url = "https://www.google.com/recaptcha/api/siteverify"

    # Параметры запроса
    params = {
        'secret': secret_key,
        'response': recaptcha_response
    }

    # Отправляем POST-запрос к Google для проверки reCAPTCHA
    response = requests.post(verify_url, data=params)

    # Получаем результат в формате JSON
    result = response.json()

    # Проверяем, успешна ли проверка
    if result['success']:
        print("reCAPTCHA успешно прошла проверку.")
        return True
    else:
        print("reCAPTCHA не прошла проверку. Причина:", result['error-codes'])
        return False

# Пример использования функции
recaptcha_response = input("Введите ответ reCAPTCHA: ")
secret_key = "6Lf9KXgUAAAAABiiJRjkSlRT0GuQHNzNM_4rRtD_"

verify_recaptcha(recaptcha_response, secret_key)
```
