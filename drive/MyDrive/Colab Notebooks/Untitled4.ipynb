```python
import requests
from bs4 import BeautifulSoup

# Функция для получения HTML-страницы по URL
def fetch_page(url):
    response = requests.get(url)
    return response.text

# Функция для извлечения информации о фильмах с HTML-страницы
def parse_movies(html):
    soup = BeautifulSoup(html, 'html.parser')
    movie_elements = soup.find_all('div', class_='desktop-rating-selection-film-item')

    movies = []

    for movie_element in movie_elements:
        title = movie_element.find('p', class_='selection-film-item-meta__name').text
        year = int(movie_element.find('p', class_='selection-film-item-meta__original-name').text[-4:])
        kp_rating_element = movie_element.find('span', class_='rating__value_positive')
        kp_rating = float(kp_rating_element.text) if kp_rating_element else None
        imdb_rating_element = movie_element.find('span', class_='rating__value')
        imdb_rating = float(imdb_rating_element.text) if imdb_rating_element else None

        movies.append({
            'title': title,
            'year': year,
            'kp_rating': kp_rating,
            'imdb_rating': imdb_rating
        })

    return movies

# URL страницы с топ-фильмами на КиноПоиск
url = 'https://www.kinopoisk.ru/lists/top250/'

# Получение и разбор данных
html = fetch_page(url)
movies = parse_movies(html)

print(movies)
```
