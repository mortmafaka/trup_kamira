6. –í —Ç–∞–±–ª–∏—Ü–µ data_temp_new —Å–æ–∑–¥–∞–¥–∏–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –ø–µ—Ä–≤–æ–≥–æ –∏ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ –∑–∞–º–µ—Ä–∞.

```python
explore_data(data_arc_new)
data_arc_new['–ù–∞—á–∞–ª–æ –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π'] = pd.to_datetime(data_arc_new['–ù–∞—á–∞–ª–æ –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π'])
data_arc_new['–ö–æ–Ω–µ—Ü –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π'] = pd.to_datetime(data_arc_new['–ö–æ–Ω–µ—Ü –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π'])
data_arc_new['–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –Ω–∞–≥—Ä–µ–≤–∞'] = (data_arc_new['–ö–æ–Ω–µ—Ü –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π'] - data_arc_new['–ù–∞—á–∞–ª–æ –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π']).dt.total_seconds()

explore_data(data_arc_new)
print(50*'=')
```

–°–æ–∑–¥–∞–¥–∏–º –ø—Ä–∏–∑–Ω–∞–∫–∏ "–ü–æ–ª–Ω–∞—è –º–æ—â–Ω–æ—Å—Ç—å" –∏ "–§–∞–∫—Ç–æ—Ä –º–æ—â–Ω–æ—Å—Ç–∏" –∏ —ç–Ω–µ—Ä–≥–∏—è —ç–ª–µ–∫—Ç—Ä–æ–¥–∞.

```python
data_arc_new['–ü–æ–ª–Ω–∞—è –º–æ—â–Ω–æ—Å—Ç—å'] = data_arc_new['–ê–∫—Ç–∏–≤–Ω–∞—è –º–æ—â–Ω–æ—Å—Ç—å'] + data_arc_new['–†–µ–∞–∫—Ç–∏–≤–Ω–∞—è –º–æ—â–Ω–æ—Å—Ç—å']
data_arc_new['–≠–Ω–µ—Ä–≥–∏—è —ç–ª–µ–∫—Ç—Ä–æ–¥–∞'] = data_arc_new['–ü–æ–ª–Ω–∞—è –º–æ—â–Ω–æ—Å—Ç—å'] * data_arc_new['–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –Ω–∞–≥—Ä–µ–≤–∞']
explore_data(data_arc_new)
```

–°–æ–∑–¥–∞–¥–∏–º –ø—Ä–∏–∑–Ω–∞–∫–∏ –∫–æ–ª–∏—á–µ—á—Å—Ç–≤–∞ –∏—Ç–µ—Ä–∞—Ü–∏–π –Ω–∞–≥—Ä–µ–≤–∞.


```python
data_arc_grouped = data_arc_new.groupby(by = 'key', as_index = False).count()[['key','–ù–∞—á–∞–ª–æ –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π']]
data_arc_grouped.columns = ['key','–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π –Ω–∞–≥—Ä–µ–≤–∞']
data_arc_new = data_arc_new.merge(data_arc_grouped, on='key', how='left')
explore_data(data_arc_new)
```

 —É–¥–∞–ª–∏–º —Å—Ç–æ–ª–±—Ü—ã –ù–∞—á–∞–ª–æ –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π –∏ –ö–æ–Ω–µ—Ü –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π

```python
data_arc_new = data_arc_new.drop(columns=['–ù–∞—á–∞–ª–æ –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π', '–ö–æ–Ω–µ—Ü –Ω–∞–≥—Ä–µ–≤–∞ –¥—É–≥–æ–π'])
explore_data(data_arc_new)
```

–ê–≥—Ä–µ–≥–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –∫–ª—é—á—É –∏ —Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—ã–π –ø—Ä–∏–∑–Ω–∞–∫

```python
data_arc_new_agg = data_arc_new.groupby('key').agg(
    active_power=('–ê–∫—Ç–∏–≤–Ω–∞—è –º–æ—â–Ω–æ—Å—Ç—å', 'sum'),
    reactive_power=('–†–µ–∞–∫—Ç–∏–≤–Ω–∞—è –º–æ—â–Ω–æ—Å—Ç—å', 'sum'),
    heat_time=('–î–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –Ω–∞–≥—Ä–µ–≤–∞', 'sum'),
    full_power=('–ü–æ–ª–Ω–∞—è –º–æ—â–Ω–æ—Å—Ç—å', 'sum'),
    heating_count=('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π –Ω–∞–≥—Ä–µ–≤–∞', 'mean'),
    energy=('–≠–Ω–µ—Ä–≥–∏—è —ç–ª–µ–∫—Ç—Ä–æ–¥–∞', 'sum'),
).reset_index()
explore_data(data_arc_new_agg)
```

```python
%%time


model_lin_reg = LinearRegression()


param_distributions = {
    'fit_intercept': [True, False],
    'positive': [True, False]
}

best_score = 1000
best_params = None

for fit_intercept in param_distributions['fit_intercept']:
    for positive in param_distributions['positive']:

        model_lin_reg.set_params(fit_intercept=fit_intercept, positive=positive)
        scores = -cross_val_score(model_lin_reg, features_train, target_train, cv=CV_COUNT, scoring='neg_mean_absolute_error')
        mean_score = np.mean(scores)

        if mean_score < best_score:
            best_score = mean_score
            best_params = {'fit_intercept': fit_intercept, 'positive': positive}

print(f"–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã: {best_params}")
print(f"–°—Ä–µ–¥–Ω—è—è MAE –ø—Ä–∏ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ –¥–ª—è –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ = {best_score}")
```
