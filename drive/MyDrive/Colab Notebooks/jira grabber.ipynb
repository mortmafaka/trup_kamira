```python
with open("jira_tickets_and_comments.txt", "w", encoding="utf-8") as file:

    # Получение списка проектов
    projects = auth_jira.projects()

    # Проход по каждому проекту
    for project in projects:
        file.write(f"Project: {project.name}\n")

        # Получение списка тикетов для текущего проекта
        issues = auth_jira.search_issues(f"project='{project.key}'", maxResults=False)

        # Проход по каждому тикету
        for issue in issues:
            file.write(f"  Issue: {issue.key} - {issue.fields.summary}\n")
            file.write(f"  Issue Text: {issue.fields.description}\n")

            # Получение списка комментариев для текущего тикета
            comments = auth_jira.comments(issue)

            # Проход по каждому комментарию
            for comment in comments:
                file.write(f"    Comment: {comment.body}\n")

            file.write("\n")
```
