Для проведения всей предобрадотки данных по населенным пунктам создадим новый столбец "clean_locality_name"

```python
data= data.dropna(subset=['locality_name'])
```

```python
data['locality_name'] = data['locality_name'].str.replace('ё', 'е', regex=True)
```

```python
data.locality_name.sort_values().unique()
```

```python
def cleaning_names (value):
    if pd.isna(value):
        return value
    elif 'поселок городского типа имени' in value:
        clean_locality_name = value.replace('поселок городского типа имени', '')
        return clean_locality_name.strip()
    elif 'поселок городского типа' in value:
        clean_locality_name = value.replace('поселок городского типа', '')
        return clean_locality_name.strip()
    elif 'городского типа' in value:
        clean_locality_name = value.replace('городского типа', '')
        return clean_locality_name.strip()
    elif 'поселок при железнодорожной станции' in value:
        clean_locality_name = value.replace('поселок при железнодорожной станции', '')
        return clean_locality_name.strip()
    elif 'коттеджный поселок' in value:
        clean_locality_name = value.replace('коттеджный поселок', '')
        return clean_locality_name.strip()
    elif 'деревня' in value:
        clean_locality_name = value.replace('деревня', '')
        return clean_locality_name.strip()
    elif 'городской поселок' in value:
        clean_locality_name = value.replace('городской поселок', '')
        return clean_locality_name.strip()
    elif 'садовое товарищество' in value:
        clean_locality_name = value.replace('садовое товарищество', '')
        return clean_locality_name.strip()
    elif 'садоводческое некоммерческое товарищество' in value:
        clean_locality_name = value.replace('садоводческое некоммерческое товарищество', '')
        return clean_locality_name.strip()
    elif 'товарищество' in value:
        clean_locality_name = value.replace('товарищество', '')
        return clean_locality_name.strip()
    elif 'поселок станции' in value:
        clean_locality_name = value.replace('поселок станции', '')
        return clean_locality_name.strip()
    elif 'станции' in value:
        clean_locality_name = value.replace('станции', '')
        return clean_locality_name.strip()
    elif 'поселок' in value:
        clean_locality_name = value.replace('поселок', '')
        return clean_locality_name.strip()
    elif 'село' in value:
        clean_locality_name = value.replace('село', '')
        return clean_locality_name.strip()
    else:
        return value
```

```python
data['clean_locality_name'] = data['locality_name'].apply(cleaning_names)
```

### Добавление новых столбцов

Для дальнейшей работы с датафреймом нам потребудется создать новые столбцы: 
<ln> •	цена одного квадратного метра;
<ln> •	день недели публикации объявления (0 — понедельник, 1 — вторник и так далее);
<ln> •	месяц публикации объявления;
<ln> •	год публикации объявления;
<ln> •	тип этажа квартиры (значения — «первый», «последний», «другой»);
<ln> •	расстояние до центра города в километрах (переведите из м в км и округлите до целых значений).

Создадим столбец с днем недели публикации объявления

```python
data['day_of_week'] = data['first_day_exposition'].dt.day_name()
```

```python
data['day_of_week'].unique()
```

Присвоим нумерацию дням недели 
<ln> - Понедельник 'Monday' - 0
<ln> - Вторник 'Tuesday' - 1
<ln> - Среда 'Wednesday' - 2
<ln> - Четверг 'Thursday' - 3
<ln> - Пятница 'Friday' - 4
<ln> - Суббота 'Saturday' - 5
<ln> - Воскресение 'Sunday' - 6

```python
def day_num(row):
    if row['day_of_week'] == 'Monday':
        return 0
    elif row['day_of_week'] == 'Tuesday':
        return 1
    elif row['day_of_week'] == 'Wednesday':
        return 2
    elif row['day_of_week'] == 'Thursday':
        return 3
    elif row['day_of_week'] == 'Friday':
        return 4
    elif row['day_of_week'] == 'Saturday':
        return 5
    elif row['day_of_week'] == 'Sunday':
        return 6
    else:
        pass
```

```python
data['day_num'] = data.apply(day_num, axis = 1)
```

Создадим столбец с месяцем в котором было сделано объявление

```python
data['month'] = data['first_day_exposition'].dt.month
```

Создадим столбец с годом публикции объявления

```python
data['year'] = data['first_day_exposition'].dt.year
```

Определим тип этажа квартиры

```python
def floor(row):
    if row['floor'] == 1:
        return 'Первый'
    elif row['floor'] == row['floors_total']:
        return 'Последний'
    elif row['floor'] <= 0:
        pass
    return 'Другой'
```

```python
data['floor_type'] = data.apply(floor, axis = 1)
```

Проверим получившиеся значения

```python
data.floor_type.unique()
```

Для будущего рассчета кореляции необходимо присоить значениям 'Другой', 'Первый', 'Последний' номера. Пусть первый будет 1, последний - 2, а другой - 0 

```python
def floor_class(row):
    if row['floor_type'] == 'Другой':
        return 0
    elif row['floor_type'] == 'Первый':
        return 1
    elif row['floor_type'] == 'Последний':
        return 2
    else:
        pass
```

```python
data['floor_class'] = data.apply(floor_class, axis = 1)
```
