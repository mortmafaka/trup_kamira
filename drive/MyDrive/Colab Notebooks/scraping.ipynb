```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
html=urlopen('http://pythonscraping.com/pages/page1.html')
bs = BeautifulSoup(html.read(), "html.parser")
print(bs.div)
```

```python
bs = BeautifulSoup(html.read(), "lxml") # работает быстрее html.parser, нужно подключать отдельно, работает с грязным кодом html
bs = BeautifulSoup(html.read(), "html5lib") # работает как LXML, медленнее, лучше других подходит для работы с сильно засранным кодом
print(bs.div)
```

# Обработка ошибок


```python
from urllib.request import HTTPError, URLError
try:
    html = urlopen('http://pythonscraping.com/pages/page1.html')
except HTTPError as e:
    print(e)
    print('Код говна: ', e.code)
    print('мы обосрались')
except URLError as e:
    print('Произошла ошибка c урлом: ', e.reason)
    print('мы обосрались')
else:
    bs = BeautifulSoup(html.read(), "html.parser")
    print('Успех!')
    print(bs.h1)
```

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup

def GetTitle(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        return None
    try:
        bs = BeautifulSoup(html.read(),'html.parser')
        title = bs.title
    except AttributeError as e:
        return None
    return title

url = 'http://vk.com/'
GetTitle(url)
```
