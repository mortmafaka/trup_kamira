```python
import numpy as np

df = df.replace(np.nan, None)
```

```python
df['renovation'].unique()
```

```python
df['url'] = 'https://realty.ya.ru/offer/' + df['offer_id'].astype(str)
df = df.drop(columns=['closest_airport', 'time_to_airport','offer_id'])
df['rooms_total'] = df['rooms_total'].fillna(0.0)
df['kitchen_space'] = df['kitchen_space'].fillna(0.0)
median_height = df['ceiling_height'].median()
df['ceiling_height'].fillna(median_height, inplace=True)
df["predicted_price_min"].fillna(df["price"], inplace=True)
df["predicted_price_max"].fillna(df["price"], inplace=True)
df["predicted_price_value"].fillna(df["price"], inplace=True)
df['renovation'].fillna('unknown')
df['mortgage_possible'].fillna('unknown', inplace=True)
df['renovation'].fillna('unknown', inplace=True)
df = df[df["metro_name"].notna()]
df = df[df["time_to_metro"].notna()]
df['year_built'] = df['year_built'].fillna(-1.0)
df['material'].fillna('unknown', inplace=True)
df['balcony'].fillna('NONE', inplace=True)
df#Много пропусков в апартаментах, поэтому заполним их False
df['is_apart'].fillna(False, inplace=True)

```

```python
# Convert 'update_date' to datetime
df['update_date'] = pd.to_datetime(df['update_date'])

# Convert 'rooms_total' and 'year_built' to int, after filling NaN values
df['rooms_total'] = df['rooms_total'].fillna(0).astype(int)
df['year_built'] = df['year_built'].fillna(0).astype(int)

# Convert 'mortgage_possible', 'balcony', 'is_apart' to category
df['mortgage_possible'] = df['mortgage_possible'].astype('category')
df['balcony'] = df['balcony'].astype('category')
df['is_apart'] = df['is_apart'].astype('category')

# Convert 'offer_type', 'flat_type', 'renovation', 'material', 'parking', 'heating' to category
df['offer_type'] = df['offer_type'].astype('category')
df['flat_type'] = df['flat_type'].astype('category')
df['renovation'] = df['renovation'].astype('category')
df['material'] = df['material'].astype('category')
df['parking'] = df['parking'].astype('category')
df['heating'] = df['heating'].astype('category')

# Extract the floor number from 'floor' and convert to int
df['floor'] = df['floor'].str.extract('(\d+)').astype('float')

df.dtypes
```

```python
file_path = "/content/filtred realty.csv"
df.to_csv(file_path, index=False)
```

```python
# Load the new data

new_df = pd.read_csv('/content/filtred realty.csv')

# Create a new dataframe with prices in millions
new_df_millions = new_df.copy()
new_df_millions['price'] = new_df['price'] / 10**6
new_df_millions['predicted_price_min'] = new_df['predicted_price_min'] / 10**6
new_df_millions['predicted_price_max'] = new_df['predicted_price_max'] / 10**6
new_df_millions['predicted_price_value'] = new_df['predicted_price_value'] / 10**6

# Filter the dataframe based on the given conditions
filtered_df_millions = new_df_millions[(new_df_millions['price'] < 8) & (new_df_millions['area'] > 30)]

filtered_df_millions.head()
```

```python
# Filter the dataframe where 'predicted_price_value' < 'price', 'price' < 8 and 'area' > 30
filtered_df = new_df_millions[(new_df_millions['predicted_price_value'] < new_df_millions['price']) &
                              (new_df_millions['price'] < 8) &
                              (new_df_millions['area'] > 30)]
```
