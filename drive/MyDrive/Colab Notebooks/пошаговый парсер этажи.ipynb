```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd

# Function to fetch data from a specific page
def fetch_data_from_page(page_number):
    page_url = f'https://spb.etagi.com/realty/?page={page_number}'

    try:
        response = requests.get(page_url, headers=headers)
        response.encoding = 'utf-8'
        response.raise_for_status()
        html = response.text
    except requests.exceptions.RequestException as e:
        print(f"Connection error: {e}")
        return None

    # Parse the HTML content
    soup = BeautifulSoup(html, 'html.parser')
    scripts = soup.find_all('script')

    # Find and display the content of the specific script
    target_script_number = 313
    script = None

    for index, script_tag in enumerate(scripts, 1):
        if index == target_script_number:
            script = script_tag.get_text()
            break

    if script:
        print(f"Script number {target_script_number}:")
        script = script[9:]
        print(script)
    else:
        print(f"Script number {target_script_number} not found.")
        # Handle this case accordingly, e.g., return or exit the program.

    # Check if the script content is not None before loading the JSON data
    if script is not None:
        try:
            data = json.loads(script)
            return data
        except json.JSONDecodeError as e:
            print(f"Invalid JSON: {e}")
    else:
        print("JSON data not available.")
        # Handle this case accordingly, e.g., return or exit the program.
        return None

# Create an empty DataFrame to store the data
all_data = pd.DataFrame()

# Define the starting and ending page numbers (adjust as needed)
start_page = 1
end_page = 136

# Fetch data from all pages and store them in the DataFrame
for page_number in range(start_page, end_page + 1):
    print(f"Processing page: {page_number}")

    data = fetch_data_from_page(page_number)
    if data is None:
        # No more pages to fetch, exit the loop
        break

    # Create DataFrame from the 'lists.flats' values in the parsed JSON data
    flats_df = create_dataframe_from_flats(data)

    # Append the DataFrame to the 'all_data' DataFrame
    all_data = pd.concat([all_data, flats_df], ignore_index=True)

# Display the first few rows of the DataFrame containing all data
print("All Data:")
print(all_data.head())
```
